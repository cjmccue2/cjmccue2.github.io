## Economic Indicators Good as Gold

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
econ1 = pd.read_csv('https://raw.githubusercontent.com/cjmccue2/cjmccue2.github.io/main/economydataset.csv',
                  na_values = {'M1SL': '.'})
```

```python
econ = econ1.rename(columns = {'PPIACO': "PPI", "CPIAUCSL": "CPI", "M1SL": "M1", "UNRATE": "Unemployment", "FEDFUNDS": "FedFunds"})
```

```python
econ = econ.fillna({'M1':0})
```

```python
econ['DATE'] = pd.to_datetime(econ['DATE'])
econ
```

In this article, I will describe some economic indicators often looked at by economists in order to determine the overall trend the economy could be headed towards. I will also provide a combination of descriptive and interactive figures in the hopes of providing information which enables the viewer to form their own conclusion on the current state of the economy.


The dataset I will be using to conduct this analysis uses the base year 1980, and looks at economic indicators such as the Producer Price Index (PPI), Consumer Price Index (CPI), M1 money supply (M1), Unemployment rate, and the FedFunds rate. 



PPI: The PPI index can be thought of as the producers side of inflation. The PPI looks at the changing cost of inputs used in production and reflects price changes before they reach consumers.

CPI: Similar to PPI, CPI is the consumers side of inflation. It reflects the changing prices once final goods do reach the market. Economists generally regard the CPI as the measure of inflation in a country.

M1: The M1 money supply is seen as the measure of money that can easily be circulated in the economy. Forms of currency that fall under M1 include cash and deposits in checking accounts. M1 is the most liquid of money measures. M2 follows M1 + savings accounts + money market funds. M3 follows M2 + long-term deposits in banks.

Unemployment: The unemployment rate is the percent of the total labor force that is unemployed. This means in times of recession the unemployment rate goes up and in times of prosperity the unemployment rate goes back down.

FedFunds: The fed funds rate is the benchmark interest rate affecting borrowing across the economy. When the fed funds rate increases, it means people looking to get loans from a bank will have to pay a higher interest rate on that loan, and generally slows the circulation of money in an economy. The opposite is true when the fed funds rate decreases. More people will be looking to borrow money because their cost of borrowing went down, this marks an increase in the circulation of money in an economy.


```python
fig = econ.plot(x='DATE', y=['PPI', 'CPI', 'M1', 'Unemployment', 'FedFunds'], figsize=(20,10))
fig.axvspan('1980-1-1','1980-7-1', alpha=0.5, color='gray')
fig.axvspan('1981-7-1','1982-11-1', alpha=0.5, color='gray')
fig.axvspan('1990-7-1','1991-3-1', alpha=0.5, color='gray')
fig.axvspan('2001-3-1','2001-11-1', alpha=0.5, color='gray')
fig.axvspan('2007-12-1','2009-6-1', alpha=0.5, color='gray')
fig.axvspan('2020-2-1','2020-4-1', alpha=0.5, color='gray')
plt.xlim(['1979-10-01', '2019-12-01'])
plt.ylabel("1980 Base Year")
plt.ylim([0, 1050])
plt.show()
```

In the figure above, we see how each of these indicators work in relation to each other. Most notably, we can see direct impacts these indicators have while in periods of recessions, which are highlighted in gray. In a recession, Unemployment rates (red) climb sharply, this is expected. Once a recession is over the unemployment rate slowly recovers until the next one. The fed funds rate (purple) declines in times of recession. The fed, which controls the fed funds rate as part of its monetary policy will either lower or raise the fed funds rate to tighten or loosen the amount of money circulating in the economy. CPI (orange) and PPI (blue) generally don't fluctuate in times of recession. Rather, they steadily increase because they are a reactionary measure to other indicators like M1, fed funds, and unemployment rates. M1 (green) is the result of policies implemented by the fed. Since the fed can either increase or decrease the amount of money circulating in the economy, M1 reacts to monetary policies implemented by the fed. The fed can buy or sell bonds in the open market, they can change the fed funds rate, and they can provide stimulus to qualifying populus in dire situations. These reasons are generally why the M1 rate increases in times of recession. During recessions, there isn't much money being spent across the economy, and the fed utilizes monetary policy to encourage spending among the populaton. When there is too much money being printed, it will be worth less than before, which affects inflation and CPI or PPI.

```python
fig = econ.plot(x='DATE', y=['PPI', 'CPI', 'M1', 'Unemployment', 'FedFunds'], figsize=(20,10))
fig.axvspan('2020-2-1','2020-4-1', alpha=0.5, color='gray')
plt.xlim(['2019-12-01', '2021-10-01'])
plt.ylabel("1980 Base Year")
plt.show()
```

The figure above depicts the economic indicators we have been talking about from the end of 2020 until recent. As seen in the graph, covid-19 has had a ginormous impact on the M1 money supply, reaching levels more than 50 times what was seen in 1980. The sudden rise and fall in the M1 supply is directly related to stimulus imposed by the fed. As mentioned earlier, inflation occurs when there is an increase of money being printed and put in circulation. 

Think of gold. Imagine if a coal mining company stumbled across a deposit of gold, uncovering more than 50 times the total supply of all the gold in the world, what would happen to the value of gold?

Well the value of gold would drop significantly. The prices of things you could buy with gold would rise in order to meet this new supply as well.

So I pose a question to the viewer, what do you think the outcome will be on inflation in the coming years? Do you think hyperinflation is destined? Or do you think the fed took this into account while disbursing stimulus rounds?


To help you come to your own conclusion, the graph below is interactive and you can choose which indicator you would like to view. In looking at the current trends between important economic indicators, one can paint a picture hinting at where the economy could be headed next.

```python
def new_field(Field = ['PPI', 'CPI', 'M1', 'Unemployment', 'FedFunds']):
    econ.plot(x='DATE', y=Field, figsize=(15,10))
    plt.axvspan('2007-12-1','2009-6-1', alpha=0.5, color='gray')
    plt.axvspan('2020-2-1','2020-4-1', alpha=0.5, color='gray')
    plt.xlim(['2006-12-01', '2021-12-01'])
    plt.ylabel("1980 Base Year")
    plt.show()
```

```python
widgets.interact(new_field)
```
